# жӯҘй©ҹе…ӯпјҡжӘўжҹҘ V.txt е…§е®№
---
- еңЁй–Ӣе§Ӣз”ҹжҲҗеүҚпјҢи«Ӣй»һж“Ҡз•«йқўе·ҰеҒҙжӘ”жЎҲпјҢй»һй–Ӣ fontdiffuser-finetune жүҫеҲ° V.txtпјҢзўәиӘҚе…§е®№жҳҜдёҚжҳҜдҪ жғіиҰҒз”ҹжҲҗзҡ„еӯ—
- еҰӮйңҖжӣҙж”№е…§е®№пјҢжңүе…©еҖӢж–№жі•
  1. еҸҜд»ҘзӣҙжҺҘз·ЁијҜйҖҷеҖӢжӘ”жЎҲпјҢжҠҠиҰҒз”ҹжҲҗзҡ„еӯ—ијёе…ҘйҖІеҺ»пјҢдёҰи·ійҒҺжӯӨжӯҘй©ҹ
  2. еҫһеҺҹе§ӢжӘ”жЎҲиЁҲз®—еҮәе°ҡжңӘз”ҹжҲҗзҡ„еӯ—пјҢи«Ӣеҹ·иЎҢд»ҘдёӢйҖҷеҖӢ cell

![vtxt.png](https://github.com/lctung/fontdiffuser-finetune-colab/blob/main/img/vtxt.png?raw=1)

```python
# дёҚеҗҢзҜ„еңҚиҰҒиҮӘе·ұеҠ 
manucripts_text = "/content/fontdiffuser-finetune/0_written_newline.txt"
font_package_text = "/content/fontdiffuser-finetune/1_MOE4808_newline.txt"

with open(manucripts_text, "r", encoding="utf-8") as file:
    manucripts_text = file.read()
with open(font_package_text, "r", encoding="utf-8") as file:
    font_package_text = file.read()

# иЁӯе®ҡдёҚеҗҢйӣҶеҗҲ
set_manucripts_text = set(manucripts_text)
set_font_package_text = set(font_package_text)

unseen_char = set_font_package_text - set_manucripts_text

unseen_char_newline = "\n".join(unseen_char)

output_path = "/content/fontdiffuser-finetune/V.txt"
with open(output_path, "w", encoding="utf-8") as file:
    file.write(unseen_char_newline)

print(f"жҲҗеҠҹдҝ®ж”№ {output_path} е…§е®№")
```

# жӯҘй©ҹе…«пјҡдҝ®ж”№ең–жӘ”еӨ§е°Ҹ

---
е°Үең–зүҮиӘҝж•ҙзӮә 300*300 зҡ„еӨ§е°Ҹ


```python
from PIL import Image
import os
from pathlib import Path

# иЁӯе®ҡең–зүҮиіҮж–ҷеӨҫи·Ҝеҫ‘
image_folder = 'outputs_generater_character_list'  # и«Ӣе°ҮжӯӨиҷ•жӣҝжҸӣзӮәжӮЁзҡ„иіҮж–ҷеӨҫи·Ҝеҫ‘

# еҸ–еҫ—жүҖжңү PNG жӘ”жЎҲ
png_files = list(Path(image_folder).glob('*.png'))

for file_path in png_files:
    try:
        with Image.open(file_path) as img:
            # иӘҝж•ҙең–зүҮеӨ§е°ҸзӮә 300x300
            img_resized = img.resize((300, 300), Image.LANCZOS)
            # е„ІеӯҳдёҰиҰҶи“ӢеҺҹе§ӢжӘ”жЎҲ
            img_resized.save(file_path)
            print(f"е·Іиҷ•зҗҶпјҡ{file_path.name}")
    except Exception as e:
        print(f"иҷ•зҗҶ {file_path.name} жҷӮзҷјз”ҹйҢҜиӘӨпјҡ{e}")
```
